# Step 2: Dataset Collection and Preprocessing

**2.1 Download the Dataset**

```python
import pandas as pd

# Load the dataset
url = "https://raw.githubusercontent.com/jbrownlee/Datasets/master/pima-indians-diabetes.data.csv"
column_names = ["Pregnancies", "Glucose", "BloodPressure", "SkinThickness", "Insulin", "BMI", "DiabetesPedigreeFunction", "Age", "Outcome"]
data = pd.read_csv(url, names=column_names)

# Display the first 5 rows
print(data.head())
```

```python
data
```

**2.2 Preprocess the Dataset**

```python
import pandas as pd
from sklearn.model_selection import train_test_split

df=data

# Shuffle the dataset to ensure randomness
df = df.sample(frac=1, random_state=42).reset_index(drop=True)

# Split the dataset into 4 equal parts
split_size = len(df) // 4
hospital_1 = df.iloc[:split_size].reset_index(drop=True)
hospital_2 = df.iloc[split_size:split_size*2].reset_index(drop=True)
hospital_3 = df.iloc[split_size*2:split_size*3].reset_index(drop=True)
hospital_4 = df.iloc[split_size*3:].reset_index(drop=True)

# Print dataset sizes
print(f"Hospital 1 Data: {hospital_1.shape}")
print(f"Hospital 2 Data: {hospital_2.shape}")
print(f"Hospital 3 Data: {hospital_3.shape}")
print(f"Hospital 4 Data: {hospital_4.shape}")

# Save each dataset
hospital_1.to_csv("hospital_1.csv", index=False)
hospital_2.to_csv("hospital_2.csv", index=False)
hospital_3.to_csv("hospital_3.csv", index=False)
hospital_4.to_csv("hospital_4.csv", index=False)
```
